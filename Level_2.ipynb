```python
# I>: ---------------Identify the most common combinations of cuisines in the dataset.-----------
#---------------------------------------------------------------------------------------------------------------------------
```

```python
# Clean data: remove rows with missing 'Cuisines' or 'Aggregate_rating'
df = df.dropna(subset=['Cuisines', 'Aggregate rating'])
```

```python
# Ensure 'Aggregate_rating' is numeric
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
# Drop rows with invalid ratings
df = df.dropna(subset=['Aggregate rating'])
```

```python
# Normalize cuisine combinations
# -----------------------------
def normalize_cuisine_combo(cuisines):
    cuisines_list = [c.strip() for c in cuisines.split(',')]
    cuisines_list.sort()
    return ', '.join(cuisines_list)

df['Cuisine_Combo'] = df['Cuisines'].apply(normalize_cuisine_combo)
```

```python
#Top 10 Most Common Cuisine Combinations:
common_combos = df['Cuisine_Combo'].value_counts().head(10)

print("Top 10 Most Common Cuisine Combinations:\n")
print(common_combos)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(y=common_combos.index, x=common_combos.values, palette='Set2')
plt.title("Top 10 Most Common Cuisine Combinations")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combination")
plt.tight_layout()
plt.show()
```

```python
# II>: ----------------Determine if certain cuisine combinations tend to have higher ratings.------------
#---------------------------------------------------------------------------------------------------------------------------
```

```python
# Group by cuisine combination
combo_stats = df.groupby('Cuisine_Combo').agg(
    Avg_Rating=('Aggregate rating', 'mean'),
    Count=('Cuisine_Combo', 'count')
).reset_index()
```

```python
# Filter to combos with at least 10 restaurants.
popular_combos = combo_stats[combo_stats['Count'] >= 10]

# Sort by average rating
top_rated_combos = popular_combos.sort_values(by='Avg_Rating', ascending=False).head(10)

print("\n Top Rated Popular Cuisine Combinations (min 10 restaurants):\n")
print(top_rated_combos)
```

```python
# Plot: Top Rated Popular Cuisine Combinations (min 10 restaurants):
plt.figure(figsize=(10, 6))
sns.barplot(y=top_rated_combos['Cuisine_Combo'], x=top_rated_combos['Avg_Rating'], palette='coolwarm')
plt.title("Top Rated Cuisine Combinations ( 10 Restaurants)")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Combination")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
import folium
from folium.plugins import MarkerCluster
from math import radians
```

```python
# Center the map on the average location
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Add marker clusters
marker_cluster = MarkerCluster().add_to(restaurant_map)

for idx, row in df.iterrows():
    name = row.get('Restaurant Name', 'Restaurant')
    rating = row.get('Aggregate rating', 'N/A')
    popup_text = f"{name}<br>Rating: {rating}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_text
    ).add_to(marker_cluster)

# Save map to HTML
restaurant_map.save('restaurant_map.html')
print("Map saved as 'restaurant_map.html'")
```

```python
# ii>: Identify any patterns or clusters of restaurants in specific areas.
#-------------------------------------------------------------------------------------------------------------------
```

```python
# Extract coordinates
coords = df[['Latitude', 'Longitude']].to_numpy()

# Convert degrees to radians for Haversine distance
coords_rad = np.radians(coords)
```

```python
# Define DBSCAN parameters
kms_per_radian = 6371.0088
epsilon = 1.0 / kms_per_radian  # 1 km radius
db = DBSCAN(eps=epsilon, min_samples=5, algorithm='ball_tree', metric='haversine')
db.fit(coords_rad)
```

```python
# Assign cluster labels
df['Cluster'] = db.labels_

# Summary
n_clusters = len(set(db.labels_)) - (1 if -1 in db.labels_ else 0)
n_noise = list(db.labels_).count(-1)
print(f"Found {n_clusters} clusters and {n_noise} noise (outlier) points.")
```

```python
# Visualize Clusters on a Map
# -----------------------------
# Initialize the map
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Create marker clusters
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Color palette for clusters
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'cadetblue', 'pink', 'black']
default_color = 'gray'

for _, row in df.iterrows():
    cluster_id = row['Cluster']
    color = colors[cluster_id % len(colors)] if cluster_id != -1 else default_color
    popup = f"Cluster: {cluster_id}<br>Rating: {row.get('Aggregate_rating', 'N/A')}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup,
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

# Save to HTML
restaurant_map.save('restaurant_cluster_map.html')
print("Cluster map saved as 'restaurant_cluster_map.html'")
```
